```python
import matplotlib.pyplot as plt
import pandas as pd

map_keys = {"ods000z0": "diff_irr",
       "gre000z0": "glob_irr",
       "gor000za": "glob_irr_std",
       "tre200s0": "temp",
       "rre150z0": "prec_rain",
       "ure200s0": "rh_air",
       "sre000z0": "sun_dur",
       "fu3010z0": "wind_speed",
       "dkl010z0": "wind_dir"}

df = pd.read_csv('./MeteoSwiss/data_meteoSwiss_2018_2019.csv', index_col='time')
df = df[map_keys.keys()]
df.index = pd.to_datetime(df.index, format='%Y%m%d%H%M')
df.columns = [map_keys[k] for k in df.columns]
df
```

```python
# --- Data cleaning

# Dataset odd values
for col in df.columns:
    df[col] = df[col].replace('-', method='bfill').astype(float)
    
# Dataset odd values
```
